```python
import numpy as np
import pandas as pd
import plotly.express as px
df = pd.read_csv("../data/irrigation_prediction.csv")
```

```python
# Color palettes
irrigation_colors = px.colors.qualitative.Set2
water_source_colors = px.colors.qualitative.Set3
```

### *31. Irrigation Type Efficiency*

```python
print("\n31. Irrigation Method Efficiency...")

method_stats = df.groupby('Irrigation_Type', observed=False).agg(
    Avg_Water_Use=('Previous_Irrigation_mm', 'mean'),
    Success_Rate=('Irrigation_Need', lambda x: (x == 'Low').mean() * 100),
    Count=('Irrigation_Type', 'count')
).reset_index()

fig31 = px.bar(method_stats, x='Irrigation_Type', y='Avg_Water_Use',
               color='Irrigation_Type', color_discrete_sequence=irrigation_colors,
               title='<b>31. Average Water Use by Irrigation Method</b>',
               text_auto='.1f',
               labels={'Avg_Water_Use': 'Avg Water Use (mm)', 'Irrigation_Type': 'Method'})

fig31.add_scatter(x=method_stats['Irrigation_Type'], y=method_stats['Success_Rate'],
                  mode='markers+lines', name='Success Rate', 
                  marker=dict(size=15, color='red'),
                  yaxis='y2')

fig31.update_layout(height=500, plot_bgcolor='white', 
                   yaxis2=dict(title='Success Rate (%)', overlaying='y', side='right'))
fig31.show()
```

### *32. Water Source Reliability*

```python
print("\n32. Water Source Reliability...")

fig32 = px.box(df, x='Water_Source', y='Previous_Irrigation_mm',
               color='Season', title='<b>32. Water Availability by Source & Season</b>',
               labels={'Previous_Irrigation_mm': 'Water Applied (mm)', 'Water_Source': 'Source'},
               color_discrete_sequence=water_source_colors)

fig32.update_layout(height=500, plot_bgcolor='white', boxmode='group')
fig32.show()
```

### *33. Rainfall Utilization Efficiency*

```python
print("\n33. Rainfall Utilization Efficiency...")

df['Rainfall_Utilization'] = (df['Soil_Moisture'] / (df['Rainfall_mm'] + 1)) * 100

fig33 = px.scatter(df, x='Rainfall_mm', y='Soil_Moisture',
                   color='Region', trendline='ols',
                   title='<b>33. Rainfall vs Soil Moisture by Region</b>',
                   labels={'Rainfall_mm': 'Rainfall (mm)', 'Soil_Moisture': 'Soil Moisture (%)'})

fig33.update_layout(height=500, plot_bgcolor='white')
fig33.show()
```

### *34. Water Source  Soil Type Compatibility*

```python
print("\n34. Water Source-Soil Compatibility...")

compatibility = df.groupby(['Water_Source', 'Soil_Type'], observed=False).agg(
    Success_Rate=('Irrigation_Need', lambda x: (x == 'Low').mean() * 100),
    Samples=('Soil_Type', 'count')
).reset_index()

fig34 = px.bar(compatibility, x='Water_Source', y='Success_Rate',
               color='Soil_Type', barmode='group',
               title='<b>34. Water Source Effectiveness by Soil Type</b>',
               labels={'Success_Rate': 'Low Irrigation Need (%)', 'Water_Source': 'Water Source'})

fig34.update_layout(height=500, plot_bgcolor='white')
fig34.show()
```

### *35. Irrigation Scheduling Accuracy*

```python
print("\n35. Irrigation Timing Analysis...")

df['Water_Deficit'] = df['Previous_Irrigation_mm'] - df.groupby(['Crop_Type', 'Crop_Growth_Stage'])['Previous_Irrigation_mm'].transform('median')

fig35 = px.scatter(df, x='Crop_Growth_Stage', y='Water_Deficit',
                   color='Irrigation_Type', facet_col='Crop_Type',
                   facet_col_wrap=3, title='<b>35. Irrigation Timing Accuracy</b><br>Positive = Over-irrigation, Negative = Under-irrigation',
                   labels={'Water_Deficit': 'Water Deficit (mm)', 'Crop_Growth_Stage': 'Growth Stage'})

fig35.update_layout(height=700, plot_bgcolor='white')
fig35.show()
```

### *36. Water Quality Impact*

```python
print("\n36. Water Quality Analysis...")

fig36 = px.scatter(df, x='Electrical_Conductivity', y='Soil_Moisture',
                   color='Water_Source', size='Soil_pH',
                   title='<b>36. Water Quality Impact on Soil</b><br>Size = Soil pH, Color = Water Source',
                   labels={'Electrical_Conductivity': 'Water EC (dS/m)', 'Soil_Moisture': 'Soil Moisture (%)'},
                   color_discrete_sequence=water_source_colors)

fig36.update_layout(height=500, plot_bgcolor='white')
fig36.show()
```

### *37. Water Conservation Practices*

```python
print("\n37. Conservation Practices...")

conservation = df.groupby(['Region', 'Mulching_Used'], observed=False).agg(
    Avg_Water_Saved=('Previous_Irrigation_mm', 'mean'),
    Success_Rate=('Irrigation_Need', lambda x: (x == 'Low').mean() * 100),
    Count=('Region', 'count')
).reset_index()

fig37 = px.bar(conservation, x='Region', y='Avg_Water_Saved',
               color='Mulching_Used', barmode='group',
               title='<b>37. Water Conservation with Mulching by Region</b>',
               labels={'Avg_Water_Saved': 'Avg Water Use (mm)', 'Region': 'Region'})

fig37.update_layout(height=500, plot_bgcolor='white')
fig37.show()
```

### *38. Water Balance Analysis*

```python
print("\n38. Water Balance Analysis...")

df['Water_Balance'] = df['Rainfall_mm'] + df['Previous_Irrigation_mm'] - df['Soil_Moisture']/2

fig38 = px.violin(df, x='Irrigation_Type', y='Water_Balance',
                  color='Crop_Type', box=True,
                  title='<b>38. Water Balance by Irrigation Method</b>',
                  labels={'Water_Balance': 'Water Balance (mm)', 'Irrigation_Type': 'Irrigation Method'})

fig38.update_layout(height=500, plot_bgcolor='white')
fig38.show()
```

### *39. Water Allocation Patterns*

```python
print("\n39. Water Allocation Patterns...")

allocation = df.groupby(['Water_Source', 'Crop_Type'], observed=False).size().reset_index(name='Count')
allocation['Percentage'] = allocation.groupby('Water_Source')['Count'].transform(lambda x: x/x.sum()*100)

fig39 = px.sunburst(allocation, path=['Water_Source', 'Crop_Type'], values='Count',
                    title='<b>39. Water Allocation: Source  Crop Distribution</b>',
                    color='Percentage', color_continuous_scale='Blues')

fig39.update_layout(height=600, plot_bgcolor='white')
fig39.show()
```

### *40. Water Use Efficiency Trends*

```python
print("\n40. Water Use Efficiency Trends...")

# Create efficiency metric
df['Water_Efficiency'] = df['Soil_Moisture'] / (df['Previous_Irrigation_mm'] + 1)

# Sort by date proxy (assume data has temporal order)
df_sorted = df.sort_values('Soil_Moisture').reset_index(drop=True)
df_sorted['Sequence'] = range(len(df_sorted))

# Calculate moving average
window = min(100, len(df_sorted)//10)
df_sorted['Efficiency_MA'] = df_sorted['Water_Efficiency'].rolling(window=window, center=True).mean()

fig40 = px.line(df_sorted, x='Sequence', y='Efficiency_MA',
                color='Irrigation_Type', title='<b>40. Water Efficiency Trends</b>',
                labels={'Efficiency_MA': 'Water Efficiency (Soil Moisture/mm)', 'Sequence': 'Time Sequence'},
                color_discrete_sequence=irrigation_colors)

fig40.update_layout(height=500, plot_bgcolor='white')
fig40.show()
```

### *41. Water Stress Indicators*

```python
print("\n41. Water Stress Analysis...")

df['Water_Stress'] = np.where(df['Soil_Moisture'] < 30, 'High',
                              np.where(df['Soil_Moisture'] < 50, 'Medium', 'Low'))

stress_matrix = df.groupby(['Crop_Type', 'Water_Stress'], observed=False).size().reset_index(name='Count')
stress_matrix['Percentage'] = stress_matrix.groupby('Crop_Type')['Count'].transform(lambda x: x/x.sum()*100)

fig41 = px.bar(stress_matrix, x='Crop_Type', y='Percentage',
               color='Water_Stress', barmode='stack',
               title='<b>41. Water Stress Levels by Crop Type</b>',
               labels={'Percentage': 'Percentage of Cases (%)', 'Crop_Type': 'Crop'},
               color_discrete_sequence=['green', 'yellow', 'red'])

fig41.update_layout(height=500, plot_bgcolor='white')
fig41.show()
```

### *42. Water Recycling Potential*

```python
print("\n42. Water Recycling Potential...")

recycling_potential = df.groupby(['Water_Source', 'Crop_Type'], observed=False).agg(
    Avg_EC=('Electrical_Conductivity', 'mean'),
    Avg_pH=('Soil_pH', 'mean'),
    Count=('Crop_Type', 'count')
).reset_index()

recycling_potential['Recycle_Safety'] = np.where(
    (recycling_potential['Avg_EC'] < 1) & (recycling_potential['Avg_pH'].between(6, 8)),
    'Safe', 'Risky'
)

fig42 = px.scatter(recycling_potential, x='Avg_EC', y='Avg_pH',
                   color='Recycle_Safety', size='Count',
                   facet_col='Crop_Type', facet_col_wrap=3,
                   title='<b>42. Water Recycling Safety Assessment</b>',
                   labels={'Avg_EC': 'Avg EC (dS/m)', 'Avg_pH': 'Avg Soil pH'},
                   color_discrete_map={'Safe': 'green', 'Risky': 'red'})

fig42.update_layout(height=700, plot_bgcolor='white')
fig42.show()
```

### *43. Energy Consumption Analysis*

```python
print("\n43. Energy-Water Nexus...")

# Assign energy factors to irrigation types
energy_factors = {'Drip': 0.8, 'Sprinkler': 1.2, 'Canal': 0.3, 'Rainfed': 0}
df['Energy_Factor'] = df['Irrigation_Type'].map(energy_factors)
df['Energy_Use'] = df['Previous_Irrigation_mm'] * df['Energy_Factor']

energy_stats = df.groupby(['Region', 'Irrigation_Type'], observed=False).agg(
    Avg_Energy=('Energy_Use', 'mean'),
    Avg_Water=('Previous_Irrigation_mm', 'mean')
).reset_index()

fig43 = px.scatter(energy_stats, x='Avg_Water', y='Avg_Energy',
                   color='Irrigation_Type', size='Avg_Energy',
                   facet_col='Region', facet_col_wrap=2,
                   title='<b>43. Energy Consumption vs Water Use</b>',
                   labels={'Avg_Water': 'Avg Water Use (mm)', 'Avg_Energy': 'Energy Index'},
                   color_discrete_sequence=irrigation_colors)

fig43.update_layout(height=600, plot_bgcolor='white')
fig43.show()
```

### *44. Water Infrastructure Analysis*

```python
print("\n44. Water Infrastructure Analysis...")

infrastructure = df.groupby('Region', observed=False).agg({
    'Irrigation_Type': lambda x: (x.isin(['Drip', 'Sprinkler'])).mean() * 100,  # % modern
    'Mulching_Used': lambda x: (x == 'Yes').mean() * 100,
    'Water_Source': lambda x: (x.isin(['Reservoir', 'River'])).mean() * 100  # % surface water
}).reset_index()

fig44 = px.bar(infrastructure.melt(id_vars='Region'), 
               x='Region', y='value', color='variable',
               barmode='group', title='<b>44. Water Infrastructure by Region</b>',
               labels={'value': 'Percentage (%)', 'variable': 'Infrastructure Metric'},
               color_discrete_sequence=px.colors.qualitative.Pastel)

fig44.update_layout(height=500, plot_bgcolor='white')
fig44.show()
```

### *45. Water Conflict Risk*

```python
print("\n45. Water Scarcity Risk...")

scarcity_risk = df.groupby('Region', observed=False).agg(
    Water_Deficit=('Water_Balance', lambda x: (x < 0).mean() * 100),
    High_Stress=('Water_Stress', lambda x: (x == 'High').mean() * 100),
    Modern_Methods=('Irrigation_Type', lambda x: (x.isin(['Drip', 'Sprinkler'])).mean() * 100)
).reset_index()

region_colors = {
    'North': '#1f77b4',
    'South': '#ff7f0e',
    'East': '#2ca02c',
    'West': '#d62728',
    'Central': '#9467bd'
}
fig45 = px.scatter_3d(
    scarcity_risk,
    x='Water_Deficit', y='High_Stress', z='Modern_Methods',
    color='Region',
    color_discrete_map=region_colors,
    size='Water_Deficit',
    title='<b>45. Water Scarcity Risk Assessment</b><br>X=Water Deficit, Y=High Stress, Z=Modern Methods',
    labels={'Water_Deficit': 'Water Deficit %',
            'High_Stress': 'High Stress %',
            'Modern_Methods': 'Modern Methods %'}
)


fig45.update_layout(height=600, plot_bgcolor='white')
fig45.update_traces(marker=dict(line=dict(color='black', width=1)))
fig45.show()
```
