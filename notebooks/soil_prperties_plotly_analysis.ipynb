```python
import numpy as np
import pandas as pd
import plotly.express as px
df = pd.read_csv("../data/irrigation_prediction.csv")
```

```python
# Display column names to see exact format
print("Column names in dataset:")
print(df.columns.tolist())
print(f"\nNumber of rows: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
```

```python
# Ensure categorical columns are properly typed
categorical_cols = ['Soil_Type', 'Crop_Type', 'Crop_Growth_Stage', 'Season', 
                    'Irrigation_Type', 'Water_Source', 'Mulching_Used', 'Region', 
                    'Irrigation_Need']

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
    else:
        print(f"Warning: Column '{col}' not found in dataset")
```

```python
# Create a color palette
soil_color_map = {'Clay': '#8B4513', 'Silt': '#D2B48C', 'Sandy': '#F4A460', 'Loamy': '#556B2F'}
irrigation_color_map = {'Low': '#2E8B57', 'Medium': '#FFA500', 'High': '#DC143C'}
region_color_map = {'North': '#1f77b4', 'South': '#ff7f0e', 'East': '#2ca02c', 
                    'West': '#d62728', 'Central': '#9467bd'}
crop_color_map = {'Wheat': '#FFD700', 'Maize': '#32CD32', 'Cotton': '#FFFFFF', 
                  'Rice': '#87CEEB', 'Sugarcane': '#D2691E', 'Potato': '#8B0000'}
```

### *1. Soil Type vs Soil Moisture*

### *2. Soil pH vs Organic Carbon*

```python
fig2 = px.scatter(df, x='Soil_pH', y='Organic_Carbon', color='Soil_Type',
                  color_discrete_map=soil_color_map, trendline='ols',
                  title='<b>2. Soil pH vs Organic Carbon Relationship</b>',
                  labels={'Soil_pH': 'Soil pH', 'Organic_Carbon': 'Organic Carbon (% mg/L)'})

fig2.update_layout(height=500, plot_bgcolor='white')
fig2.show()
```

### *3. Electrical Conductivity Distribution*

```python
print("\n3. Soil Salinity Analysis...")

fig3 = px.violin(df, x='Soil_Type', y='Electrical_Conductivity', color='Region',
                 box=True, points='outliers',
                 title='<b>3. Soil Salinity (EC) by Soil Type & Region</b>',
                 labels={'Electrical_Conductivity': 'EC (dS/m)', 'Soil_Type': 'Soil Type'})

fig3.update_layout(height=500, plot_bgcolor='white')
fig3.show()
```

### *4. Organic Carbon vs Yield Potential*

### *5. Soil pH Distribution by Region*

### *6. EC by Irrigation Type & Water Source*

```python
print("\n6. Salinity by Water Source...")

fig6 = px.box(df, x='Irrigation_Type', y='Electrical_Conductivity', color='Water_Source',
              title='<b>6. Soil Salinity by Irrigation Method & Water Source</b>',
              labels={'Electrical_Conductivity': 'EC (dS/m)', 'Irrigation_Type': 'Irrigation Method'})

fig6.update_layout(height=500, plot_bgcolor='black', boxmode='group')
fig6.show()
```

### *7. Mulching Impact on Soil Moisture*

```python
print("\n7. Mulching Benefits...")

mulching_impact = df.groupby(['Soil_Type', 'Mulching_Used'], observed=False).agg(
    Avg_Moisture=('Soil_Moisture', 'mean'),
    Count=('Soil_Type', 'count')
).reset_index()

fig7 = px.bar(mulching_impact, x='Soil_Type', y='Avg_Moisture', color='Mulching_Used',
              barmode='group', text_auto='.1f',
              title='<b>7. Mulching Impact on Soil Moisture Retention</b>',
              labels={'Avg_Moisture': 'Avg Soil Moisture (%)', 'Soil_Type': 'Soil Type'})

fig7.update_layout(height=500, plot_bgcolor='white')
fig7.show()
```

### *8. Soil Health Radar Chart*

```python
import plotly.graph_objects as go

print("\n8. Soil Health Assessment...")

# Calculate soil health scores
soil_health = df.groupby('Soil_Type', observed=False).agg({
    'Soil_pH': lambda x: (100 - abs(x.mean() - 6.5) * 10).clip(0, 100),
    'Organic_Carbon': lambda x: min(100, x.mean() * 25),
    'Electrical_Conductivity': lambda x: max(0, 100 - x.mean() * 20),
    'Soil_Moisture': 'mean'
}).reset_index()

# Normalize moisture to 0-100 scale
soil_health['Soil_Moisture'] = soil_health['Soil_Moisture'].clip(0, 100)

fig8 = go.Figure()
categories = ['pH Balance', 'Organic Carbon', 'Low Salinity', 'Moisture']

for soil_type in soil_color_map.keys():
    if soil_type in soil_health['Soil_Type'].values:
        scores = soil_health[soil_health['Soil_Type'] == soil_type].iloc[0][1:].tolist()
        scores.append(scores[0])  # Close the radar
        fig8.add_trace(go.Scatterpolar(r=scores, theta=categories + [categories[0]],
                                       name=soil_type, fill='toself',
                                       line_color=soil_color_map[soil_type]))

fig8.update_layout(polar=dict(radialaxis=dict(range=[0, 100])),
                   title='<b>8. Soil Health Radar Chart</b>', height=500)
fig8.show()
```

### *9. Clay Soil Correlation Matrix*

```python
print("\n9. Clay Soil Correlations...")

clay_df = df[df['Soil_Type'] == 'Clay']
numeric_cols = ['Soil_pH', 'Soil_Moisture', 'Organic_Carbon', 
                'Electrical_Conductivity', 'Rainfall_mm', 'Previous_Irrigation_mm']
corr_matrix = clay_df[numeric_cols].corr()

fig9 = px.imshow(corr_matrix, text_auto='.2f', color_continuous_scale='RdBu_r',
                 title='<b>9. Clay Soil: Feature Correlation Matrix</b>',
                 labels={'color': 'Correlation'})

fig9.update_layout(height=500)
fig9.show()
```

### *10. Soil Degradation Risk*

```python
# Soil Degradation Risk: pH vs Salinity

print("\n10. Degradation Risk...")

df['Degradation_Risk'] = (
    (df['Electrical_Conductivity'] > 2).astype(int) * 40 +
    ((df['Soil_pH'] < 5.5) | (df['Soil_pH'] > 8.5)).astype(int) * 30 +
    (df['Organic_Carbon'] < 0.8).astype(int) * 20 +
    (df['Soil_Moisture'] < 25).astype(int) * 10
)

fig10 = px.scatter(df, x='Soil_pH', y='Electrical_Conductivity',
                   color='Degradation_Risk', size='Degradation_Risk',
                   color_continuous_scale='RdYlGn_r',
                   title='<b>10. Soil Degradation Risk: pH vs Salinity</b>',
                   labels={'Electrical_Conductivity': 'EC (dS/m)', 'Soil_pH': 'Soil pH'})

fig10.update_layout(height=500, plot_bgcolor='black')
fig10.show()

# Soil Degradation Risk by Soil Type
print("\n10b. Soil Degradation Risk by Soil Type...")

fig10b = px.box(df, x='Soil_Type', y='Degradation_Risk',
               title='<b>10b. Soil Degradation Risk by Soil Type</b>',
               labels={'Degradation_Risk': 'Degradation Risk Score'})

fig10b.update_layout(height=500, plot_bgcolor='white')
fig10b.show()
```

### *12. Organic Carbon Trends*

```python
print("\n12. Organic Carbon Trends...")

# Sort by field size as proxy for time/management
df_sorted = df.sort_values('Field_Area_hectare').reset_index(drop=True)
df_sorted['Sequence'] = range(len(df_sorted))
df_sorted['OC_MA'] = df_sorted['Organic_Carbon'].rolling(100, center=True).mean()

fig12 = px.line(df_sorted, x='Sequence', y='OC_MA', color='Mulching_Used',
                title='<b>12. Organic Carbon Trends (Moving Average)</b>',
                labels={'OC_MA': 'Organic Carbon (% mg/L)', 'Sequence': 'Sample Sequence'})

fig12.update_layout(height=500, plot_bgcolor='white')
fig12.show()
```

### *13. Soil Compaction Indicators*

```python
print("\n13. Compaction Risk...")

df['Compaction_Risk'] = (
    df['Soil_Type'].map({'Clay': 0.4, 'Silt': 0.3, 'Loamy': 0.2, 'Sandy': 0.1}) +
    (df['Irrigation_Type'].isin(['Canal', 'Rainfed'])).astype(int) * 0.3 +
    (df['Soil_Moisture'] > 70).astype(int) * 0.3
) * 100

sample = df.sample(300)

soil_order = list(soil_color_map.keys())
soil_map = {s: i for i, s in enumerate(soil_order)}

sample = sample.assign(
    Soil_Type_Code=sample["Soil_Type"].map(soil_map)
)

fig13 = px.parallel_coordinates(
    sample,
    dimensions=[
        "Soil_Moisture",
        "Organic_Carbon",
        "Electrical_Conductivity",
        "Compaction_Risk",
    ],
    color="Soil_Type_Code",
    color_continuous_scale=[soil_color_map[s] for s in soil_order],
    range_color=[0, len(soil_order) - 1],
    title="<b>13. Soil Compaction Risk Profile</b>",
)

fig13.update_coloraxes(
    colorbar=dict(
        tickmode="array",
        tickvals=list(range(len(soil_order))),
        ticktext=soil_order,
    )
)

fig13.update_layout(height=500, plot_bgcolor='white')

fig13.update_layout(height=500)
fig13.show()
```

### *14. Soil Type Suitability*

```python
print("\n14. Soil Suitability...")

suitability = df.groupby(['Soil_Type', 'Crop_Type'], observed=False).agg(
    Success_Rate=('Irrigation_Need', lambda x: (x == 'Low').mean() * 100),
    Count=('Soil_Type', 'count')
).reset_index()

fig14 = px.treemap(suitability, path=['Soil_Type', 'Crop_Type'], values='Count',
                   color='Success_Rate', color_continuous_scale='RdYlGn',
                   title='<b>14. Soil Type Suitability for Crops</b><br>Color = Success Rate')

fig14.update_layout(height=500)
fig14.show()
```

```python
print("\nCreating Plot 14b: Soil Type Suitability Analysis...")

# Calculate success rate (Low Irrigation Need) for each combination
success_data = []
for region in df['Region'].cat.categories:
    region_df = df[df['Region'] == region]
    for soil_type in soil_color_map.keys():
        if soil_type in df['Soil_Type'].cat.categories:
            soil_df = region_df[region_df['Soil_Type'] == soil_type]
            for crop in df['Crop_Type'].cat.categories:
                crop_df = soil_df[soil_df['Crop_Type'] == crop]
                if len(crop_df) > 0:
                    success_rate = (crop_df['Irrigation_Need'] == 'Low').mean() * 100
                    avg_moisture = crop_df['Soil_Moisture'].mean()
                    success_data.append({
                        'Region': region,
                        'Soil_Type': soil_type,
                        'Crop_Type': crop,
                        'Success_Rate': success_rate,
                        'Avg_Moisture': avg_moisture,
                        'Sample_Count': len(crop_df)
                    })

if success_data:
    success_df = pd.DataFrame(success_data)
    
    fig14b = px.treemap(success_df, 
                      path=['Region', 'Soil_Type', 'Crop_Type'],
                      values='Sample_Count',
                      color='Success_Rate',
                      color_continuous_scale='RdYlGn',
                      range_color=[0, 100],
                      hover_data=['Success_Rate', 'Avg_Moisture', 'Sample_Count'],
                      title='<b>14b. Soil Type Suitability Analysis</b><br><sup>Color = Success Rate (Low Irrigation Need %) | Size = Sample Count</sup>',
                      height=700)
    
    fig14b.update_layout(
        coloraxis_colorbar=dict(
            title="Success Rate %",
            tickformat=".0f",
            ticksuffix="%",
            thickness=20
        ),
        plot_bgcolor='black'
    )
    
    print(" Plot 14b created successfully")
    fig14b.show()
else:
    print(" No success data available for treemap")
```

### *15. Soil Amendment Needs*

```python
print("\n15. Amendment Requirements...")

amendments = df.groupby('Soil_Type', observed=False).agg(
    pH_Needs=('Soil_pH', lambda x: ((x < 5.5) | (x > 8.5)).mean() * 100),
    OC_Needs=('Organic_Carbon', lambda x: (x < 0.8).mean() * 100),
    EC_Needs=('Electrical_Conductivity', lambda x: (x > 2).mean() * 100)
).reset_index()

fig15 = px.bar(amendments.melt(id_vars='Soil_Type'), 
               x='Soil_Type', y='value', color='variable',
               barmode='group', title='<b>15. Soil Amendment Needs by Soil Type</b>',
               labels={'value': 'Percentage Needing (%)', 'variable': 'Amendment Type'},
               color_discrete_sequence=['red', 'green', 'orange'])

fig15.update_layout(height=500, plot_bgcolor='black')
fig15.show()
```
